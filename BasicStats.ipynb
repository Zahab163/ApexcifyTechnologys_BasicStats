##Descriptive Statistics on California Housing Dataset

Task : Basic Descriptive Statistics on Data Set
Objective:
- Load any small Dataset
_ Calculate descriptive statistics : mean,median min ,max standard deviation .
- Tools : Python, Pandas
- Key Concept : describe(), mean(), std(), min() , max()
- I took this dataset from colab 's sample_data folder

```python
import pandas as pd
import numpy as np

# Load dataset directly from Colab sample_data
df = pd.read_csv("/content/sample_data/california_housing_train.csv")



```

```python
# Preview first rows
print("Dataset Preview:")
df.head()
```

```python
# Descriptive statistics using describe()
print("\nDescriptive Statistics:")
print(df.describe())
```

### Calculate specific statistics for one column (e.g., median_income)

```python
# Find mean
print("\nMean of median_income:", df["median_income"].mean())
```

```python
# Find median
print("Median of median_income:", df["median_income"].median())
```

```python
# Find Min
print("Minimum of median_income:", df["median_income"].min())
```

```python
#Find Max
print("Maximum of median_income:", df["median_income"].max())
```

```python
# Find Standard Deviation
print("Standard Deviation of median_income:", df["median_income"].std())
```

##Key Concepts
- read_csv() → load dataset.
- describe() → quick overview of all numeric columns.
- mean(), median(), min(), max(), std() → specific descriptive statistics
